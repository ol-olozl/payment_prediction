```python
# 1. `visit_month` 컬럼을 기준으로 그룹화하고 각 월별 방문 횟수를 계산합니다.
monthly_total_visits = feature_attended_df['visit_month'].value_counts().sort_index()

# 2. 월별 전체 방문 횟수를 막대 그래프로 시각화합니다.
plt.figure(figsize=(12, 6))
sns.barplot(x=monthly_total_visits.index, y=monthly_total_visits.values, palette='viridis')
plt.title('월별 전체 방문 횟수 분포')
plt.xlabel('방문 월')
plt.ylabel('방문 횟수')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--')
plt.show()

# 3. `feature_attended_df` 데이터프레임을 `visit_month`와 `is_payment` 컬럼을 기준으로 그룹화하고 각 조합별 방문 횟수를 계산합니다.
monthly_payment_visits = feature_attended_df.groupby(['visit_month', 'is_payment']).size().unstack(fill_value=0)

# 4. `visit_month`별 `is_payment` 여부에 따른 방문 횟수를 누적 막대 그래프로 시각화합니다.
monthly_payment_visits.plot(kind='bar', stacked=True, figsize=(12, 7), colormap='viridis')
plt.title('월별 결제 여부에 따른 방문 횟수')
plt.xlabel('방문 월')
plt.ylabel('방문 횟수')
plt.xticks(rotation=0)
plt.legend(title='결제 여부', labels=['미결제 (0)', '결제 (1)'])
plt.grid(axis='y', linestyle='--')
plt.show()
```

```python
# 1. 각 월별 전체 방문 횟수 (결제 + 미결제)를 계산합니다.
# monthly_total_visits 변수는 이전 단계에서 이미 계산되었습니다.

# 2. 각 월별 결제 전환율을 계산합니다.
# monthly_payment_visits DataFrame에서 각 월의 결제 건수 (is_payment=1)를 가져옵니다.
monthly_paid_visits = monthly_payment_visits[1]

# 결제 전환율 계산: (해당 월의 결제 건수 / 해당 월의 전체 방문 횟수) * 100
monthly_conversion_rate = (monthly_paid_visits / monthly_total_visits) * 100

# 계산된 월별 결제 전환율을 DataFrame으로 변환 (시각화를 위해)
monthly_conversion_rate_df = monthly_conversion_rate.reset_index()
monthly_conversion_rate_df.columns = ['visit_month', 'conversion_rate']


# 3. 계산된 월별 결제 전환율을 막대 그래프로 시각화합니다.
plt.figure(figsize=(12, 6))
sns.barplot(x='visit_month', y='conversion_rate', data=monthly_conversion_rate_df, palette='viridis')

# 4. 그래프에 적절한 제목, x축 레이블, y축 레이블을 추가합니다.
plt.title('월별 결제 전환율')
plt.xlabel('방문 월')
plt.ylabel('결제 전환율 (%)')

# 5. 필요하다면 y축 범위를 조정하여 전환율 차이를 더 잘 드러나게 합니다.
# 전환율 범위에 따라 조정 가능 (예: plt.ylim(10, 30))
# 현재 데이터에서 큰 변동은 없어 보이나, 필요시 주석 해제 후 사용
# plt.ylim(15, 35)

# x축 눈금 회전 방지
plt.xticks(rotation=0)

# y축 그리드 추가
plt.grid(axis='y', linestyle='--')

# 6. 그래프를 표시합니다.
plt.show()
```

```python
df_stay_time = df[['user_uuid', 'stay_time_second']]
```

```python
q1 = df_stay_time['stay_time_second'].quantile(0.25)
q3 = df_stay_time['stay_time_second'].quantile(0.75)
iqr = q3 - q1

lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

df_stay_time = df_stay_time[(df_stay_time['stay_time_second'] >= lower_bound) & (df_stay_time['stay_time_second'] <= upper_bound)]
```

```python
sns.boxplot(df_stay_time['stay_time_second'])
```

```python
df_stay_pay = df_stay_time.merge(trial_payment, on = 'user_uuid', how = 'left')
```

```python
df_stay_pay
```

```python
df_stay_pay['stay_bin'] = pd.qcut(df_stay_pay['stay_time_second'], q=5)

grouped = df_stay_pay.groupby('stay_bin')['is_payment'].agg(['count', 'sum'])
grouped['payment_ratio'] = grouped['sum'] / grouped['count']
grouped = grouped.reset_index()

plt.figure(figsize=(10,6))
sns.barplot(x='stay_bin', y='payment_ratio', data=grouped)
plt.xticks(rotation=45)
plt.ylabel('Payment Ratio')
plt.title('Payment Ratio by Stay Time (Binned)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 합성 접속 기록을 저장할 빈 리스트 생성
synthetic_access_records = []

# missing_users_visit_info_df의 각 행을 순회하며 합성 기록 생성
for index, row in missing_users_visit_info_df.iterrows():
    user_uuid = row['user_uuid']
    site_id = row['site_id'] # visit_info의 site_id 사용
    first_enter_time = row['first_enter_time']
    last_leave_time = row['last_leave_time']
    # access_count는 나중에 전체 데이터 기준으로 다시 계산할 것이므로 임시값 또는 None 할당
    # id도 나중에 재정렬하거나 고유값 생성 필요

    # 최초 입실 기록 생성
    synthetic_access_records.append({
        'user_uuid': user_uuid,
        'cdate': first_enter_time,
        'checkin': 1,
        'site_id': site_id,
        'id': None, # 임시 ID 또는 None
        'access_count': None # 나중에 다시 계산
    })

    # 최후 퇴실 기록 생성 (last_leave_time이 first_enter_time과 다른 경우에만)
    if first_enter_time != last_leave_time:
         synthetic_access_records.append({
            'user_uuid': user_uuid,
            'cdate': last_leave_time,
            'checkin': 2,
            'site_id': site_id,
            'id': None, # 임시 ID 또는 None
            'access_count': None # 나중에 다시 계산
        })
    else: # 체류 시간이 0인 경우 (first_enter_time == last_leave_time) checkin 1 기록만 생성
         synthetic_access_records.append({
            'user_uuid': user_uuid,
            'cdate': last_leave_time,
            'checkin': 1, # 체류시간 0이면 입/퇴실 시간이 같으므로 checkin 1로 통일
            'site_id': site_id,
            'id': None, # 임시 ID 또는 None
            'access_count': None # 나중에 다시 계산
        })


# 합성 기록 리스트를 데이터프레임으로 변환
synthetic_access_df = pd.DataFrame(synthetic_access_records)

# cdate_date 컬럼 추가 (cdate에서 추출)
synthetic_access_df['cdate_date'] = synthetic_access_df['cdate'].dt.date

# 결과 확인
print("생성된 합성 접속 기록 (처음 5개):")
display(synthetic_access_df.head())
print("\n합성 기록 데이터프레임 형태:")
print(synthetic_access_df.shape)
synthetic_access_df.info()
```

```python
import numpy as np
import pandas as pd
from datetime import timedelta
# 이건 너무 어려워서 GPT 도움

def mean_circular_time(times):
    """
    시간 리스트에서 원형 평균(시계방향 기준 평균 시각)을 계산해 datetime.time 형태로 반환
    :param times: list-like or pd.Series of str, datetime, or timedelta (예: '23:35:00', pd.Timestamp)
    :return: 평균 시각 (datetime.time)
    """

    # 문자열 → datetime.time 변환 (pandas가 알아서 처리)
    times = pd.to_datetime(times, format='%H:%M:%S', errors='coerce').dropna()
    if times.empty:
        return None

    # 하루 총 초
    SECONDS_IN_DAY = 24 * 60 * 60

    # 초 단위로 환산
    seconds = times.dt.hour * 3600 + times.dt.minute * 60 + times.dt.second

    # 각도로 변환
    angles = seconds / SECONDS_IN_DAY * 2 * np.pi

    # 평균 각도 구하기
    mean_angle = np.arctan2(np.mean(np.sin(angles)), np.mean(np.cos(angles)))
    if mean_angle < 0:
        mean_angle += 2 * np.pi

    # 다시 초로 변환
    mean_seconds = mean_angle / (2 * np.pi) * SECONDS_IN_DAY

    # 평균 시각으로 변환
    return (pd.Timestamp('1900-01-01') + timedelta(seconds=mean_seconds)).time()
```

```python
from datetime import datetime

def map_time_of_day(t):
    hour = t.hour if isinstance(t, (pd.Timestamp, datetime)) else t.hour

    if 0 <= hour < 6:
        return 'night'
    elif 6 <= hour < 12:
        return 'morning'
    elif 12 <= hour < 18:
        return 'afternoon'
    else:
        return 'evening'
```

```python
# 가장 자주 방문한 지점 뽑기
def get_most_visited_site(df, user_col='user_uuid', site_col='site_id'):
    most_visited = (
        df.groupby([user_col, site_col])
        .size()
        .reset_index(name='visit_count')
        .sort_values(['user_uuid', 'visit_count'], ascending=[True, False])
    )

    # 유저별 최다 방문 1개씩만 남기기
    most_visited = most_visited.drop_duplicates(subset=[user_col], keep='first')
    most_visited = most_visited[[user_col, site_col]].rename(columns={site_col: 'most_visited_site'})

    return most_visited
```

```python
# 범주형 변수 인코딩
X = pd.get_dummies(X, columns=cat_cols)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y,test_size=0.2, random_state=42)
```

```python
# 수치형 변수 스케일링
scaler = StandardScaler()
X_train_num = pd.DataFrame(scaler.fit_transform(X_train[num_cols]), columns=num_cols, index=X_train.index)
X_test_num = pd.DataFrame(scaler.transform(X_test[num_cols]), columns=num_cols, index=X_test.index)

# 수치형 + 범주형 결합
X_train_final = pd.concat([X_train_num, X_train.drop(columns=num_cols)], axis=1)
X_test_final = pd.concat([X_test_num, X_test.drop(columns=num_cols)], axis=1)
```

```python
results = {}

def evaluate_model(name, model, X_train, y_train, X_test, y_test):
    # 예측값
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)

    # 확률 예측값 (테스트용)
    y_test_proba = model.predict_proba(X_test)[:, 1]

    # 결과 저장
    results[name] = {
        "Accuracy_train": accuracy_score(y_train, y_train_pred),
        "Accuracy_test": accuracy_score(y_test, y_test_pred),
        "Precision": precision_score(y_test, y_test_pred),
        "Recall": recall_score(y_test, y_test_pred),
        "F1": f1_score(y_test, y_test_pred),
        "ROC_AUC": roc_auc_score(y_test, y_test_proba)
    }

    print(f"\nModel: {name}")
    for k, v in results[name].items():
        print(f"{k}: {v:.4f}")
```

```python
# @title Bagging 별도 처리
bagging_models = {
    "Bagging (Decision Tree)": BaggingClassifier(estimator=models["Decision Tree"], random_state=42),
    "Bagging (Decision Tree Balanced)": BaggingClassifier(estimator=models["Decision Tree Balanced"], random_state=42)
}
```

```python

# conversion_rate 시각화
plt.figure(figsize=(12, 6))
sns.barplot(x=conversion_by_bin.index, y='conversion_rate', data=conversion_by_bin)
plt.title('일 평균 체류 시간 구간별 결제율')
plt.xlabel('체류 시간 구간')
plt.ylabel('결제율 (%)')
plt.xticks(rotation=45, ha='right')
plt.ylim(0, conversion_by_bin['conversion_rate'].max() * 1.1) # y축 범위 조정
plt.grid(axis='y', linestyle='--', alpha=0.7)

# 각 막대 위에 결제율 값 표시
for index, row in conversion_by_bin.iterrows():
  plt.text(index, row['conversion_rate'], f'{row["conversion_rate"]:.1f}%', color='black', ha="center", va="bottom")

plt.tight_layout()
plt.show()
```

## visit_count 별 시각화

```python
# visit_count 1, 2, 3 별로 데이터 분할
df_visit_1 = new_df[new_df['visit_count'] == 1].copy()
df_visit_2 = new_df[new_df['visit_count'] == 2].copy()
df_visit_3 = new_df[new_df['visit_count'] == 3].copy()

# 일 평균 체류 시간 (시간 단위) 계산
df_visit_1['avg_daily_stay_hour'] = df_visit_1['total_stay_time'] / 3600 # visit_count가 1이니 일 평균과 같음
df_visit_2['avg_daily_stay_hour'] = (df_visit_2['total_stay_time'] / df_visit_2['visit_count']) / 3600
df_visit_3['avg_daily_stay_hour'] = (df_visit_3['total_stay_time'] / df_visit_3['visit_count']) / 3600

# 체류 시간 구간 정의 (위에서 사용한 구간과 동일하게)
bins = [0, 1, 2, 3, 5, 8, 12, 18, 24, 1000] # 24시간 초과를 포함하기 위해 최대값 늘림
labels = ['0~1h', '1~2h', '2~3h', '3~5h', '5~8h', '8~12h', '12~18h', '18~24h', '24h+']

# visit_count 1일 때 구간별 결제율
df_visit_1['duration_bin'] = pd.cut(df_visit_1['avg_daily_stay_hour'], bins=bins, labels=labels, right=False)
conversion_visit_1 = df_visit_1.groupby('duration_bin')['is_payment'].agg(['count', 'sum'])
conversion_visit_1['conversion_rate'] = (conversion_visit_1['sum'] / conversion_visit_1['count']) * 100
conversion_visit_1.rename(columns={'count': 'total_users', 'sum': 'payments'}, inplace=True)
conversion_visit_1 = conversion_visit_1.reindex(labels).dropna(how='all') # 모든 레이블 포함하고 데이터 없는 구간 제거

# visit_count 2일 때 구간별 결제율
df_visit_2['duration_bin'] = pd.cut(df_visit_2['avg_daily_stay_hour'], bins=bins, labels=labels, right=False)
conversion_visit_2 = df_visit_2.groupby('duration_bin')['is_payment'].agg(['count', 'sum'])
conversion_visit_2['conversion_rate'] = (conversion_visit_2['sum'] / conversion_visit_2['count']) * 100
conversion_visit_2.rename(columns={'count': 'total_users', 'sum': 'payments'}, inplace=True)
conversion_visit_2 = conversion_visit_2.reindex(labels).dropna(how='all') # 모든 레이블 포함하고 데이터 없는 구간 제거

# visit_count 3일 때 구간별 결제율
df_visit_3['duration_bin'] = pd.cut(df_visit_3['avg_daily_stay_hour'], bins=bins, labels=labels, right=False)
conversion_visit_3 = df_visit_3.groupby('duration_bin')['is_payment'].agg(['count', 'sum'])
conversion_visit_3['conversion_rate'] = (conversion_visit_3['sum'] / conversion_visit_3['count']) * 100
conversion_visit_3.rename(columns={'count': 'total_users', 'sum': 'payments'}, inplace=True)
conversion_visit_3 = conversion_visit_3.reindex(labels).dropna(how='all') # 모든 레이블 포함하고 데이터 없는 구간 제거


# 결과 출력
print("\nvisit_count = 1 일 평균 체류 시간 구간별 결제율:")
print(conversion_visit_1)
print("\nvisit_count = 2 일 평균 체류 시간 구간별 결제율:")
print(conversion_visit_2)
print("\nvisit_count = 3 일 평균 체류 시간 구간별 결제율:")
print(conversion_visit_3)


# 시각화
plt.figure(figsize=(15, 8))

# visit_count 1
plt.subplot(1, 3, 1)
sns.barplot(x=conversion_visit_1.index, y='conversion_rate', data=conversion_visit_1)
plt.title('Visit Count 1')
plt.xlabel('Avg Daily Stay Time')
plt.ylabel('Conversion Rate (%)')
plt.xticks(rotation=45, ha='right')
plt.ylim(0, max(conversion_visit_1['conversion_rate'].max() if not conversion_visit_1.empty else 0,
                conversion_visit_2['conversion_rate'].max() if not conversion_visit_2.empty else 0,
                conversion_visit_3['conversion_rate'].max() if not conversion_visit_3.empty else 0) * 1.1)
plt.grid(axis='y', linestyle='--', alpha=0.7)
for index, row in conversion_visit_1.iterrows():
    plt.text(conversion_visit_1.index.get_loc(index), row['conversion_rate'], f'{row["conversion_rate"]:.1f}%', color='black', ha="center", va="bottom")


# visit_count 2
plt.subplot(1, 3, 2)
sns.barplot(x=conversion_visit_2.index, y='conversion_rate', data=conversion_visit_2)
plt.title('Visit Count 2')
plt.xlabel('Avg Daily Stay Time')
plt.ylabel('') # y축 레이블 중복 방지
plt.xticks(rotation=45, ha='right')
plt.ylim(0, max(conversion_visit_1['conversion_rate'].max() if not conversion_visit_1.empty else 0,
                conversion_visit_2['conversion_rate'].max() if not conversion_visit_2.empty else 0,
                conversion_visit_3['conversion_rate'].max() if not conversion_visit_3.empty else 0) * 1.1)
plt.grid(axis='y', linestyle='--', alpha=0.7)
for index, row in conversion_visit_2.iterrows():
    plt.text(conversion_visit_2.index.get_loc(index), row['conversion_rate'], f'{row["conversion_rate"]:.1f}%', color='black', ha="center", va="bottom")


# visit_count 3
plt.subplot(1, 3, 3)
sns.barplot(x=conversion_visit_3.index, y='conversion_rate', data=conversion_visit_3)
plt.title('Visit Count 3')
plt.xlabel('Avg Daily Stay Time')
plt.ylabel('') # y축 레이블 중복 방지
plt.xticks(rotation=45, ha='right')
plt.ylim(0, max(conversion_visit_1['conversion_rate'].max() if not conversion_visit_1.empty else 0,
                conversion_visit_2['conversion_rate'].max() if not conversion_visit_2.empty else 0,
                conversion_visit_3['conversion_rate'].max() if not conversion_visit_3.empty else 0) * 1.1)
plt.grid(axis='y', linestyle='--', alpha=0.7)
for index, row in conversion_visit_3.iterrows():
    plt.text(conversion_visit_3.index.get_loc(index), row['conversion_rate'], f'{row["conversion_rate"]:.1f}%', color='black', ha="center", va="bottom")


plt.suptitle('Visit Count 별 일 평균 체류 시간 구간별 결제율', y=1.02, fontsize=16) # 전체 제목
plt.tight_layout()
plt.show()
```
